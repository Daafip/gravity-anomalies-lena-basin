```python
# loading other layers
outline           = gpd.read_file(f"Data\\lena_basin_outline_polygon.gpkg",driver="GPKG")
main_rivers       = gpd.read_file(f"Data\\lena_main_river.gpkg",driver="GPKG")   
coast             = gpd.read_file(f"Data\\north_east_russian_coastlines.gpkg",driver="GPKG")   
# fixing crs
for layer in [outline,main_rivers,coast]:
    layer.geometry = layer.geometry.to_crs("EPSG:4326")
```

```python
fname_filtered = "Data\\anomally_waterhead_filtered_multicore.nc" # debug
fname_raw = "Data\\anomally_waterhead_raw_multicore.nc"  # debug
```

```python
ds_filtered = xr.open_dataset(fname_filtered)
ds_raw = xr.open_dataset(fname_raw)
```

```python
mask = regionmask.mask_geopandas(outline, 
                                 ds_filtered.lon.to_numpy(), 
                                 ds_filtered.lat.to_numpy(),
                                 lon_name="lon",
                                 lat_name="lat")
```

```python
ds_filtered_basin = ds_filtered.where(mask==0)
ds_raw_basin = ds_raw.where(mask==0)
```

```python
changes_total_water_mass_filtered = ds_filtered_basin.mean(dim=["lat","lon"])["dh(m)"] * 100
changes_total_water_mass_raw = ds_raw_basin.mean(dim=["lat","lon"])["dh(m)"] * 100
```

```python
fig, ax = plt.subplots()
n = 3
changes_total_water_mass_filtered[n:].plot(ax=ax,label="Filtered")
changes_total_water_mass_raw[n:].plot(ax=ax,label="Unfiltered")
ax.set_xlabel("Date")
ax.set_ylabel("Mass anomaly   $\mu \delta h_w$ [cm]")
ax.set_title("Mass anomalies over time in the lena basin"); 
ax.legend()
```
